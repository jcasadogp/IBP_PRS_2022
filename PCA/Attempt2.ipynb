Continue from 1kG_5. Need to begin by removing ambiguous SNPs and only retaining biallelic SNPs.

```bash
#find ambiguous SNPs
cat final_IBP.bim | \
awk '!(($5=="A" && $6=="T") || \
     ($5=="T" && $6=="A") || \
     ($5=="G" && $6=="C") || \
     ($5=="C" && $6=="G")) {print}' > IBP_non_ambiguous_snps.txt 
```

```bash
plink --bfile final_IBP --extract IBP_non_ambiguous_snps.txt --allow-no-sex --make-bed --out IBP_2
```

```bash
#remove biallelic

awk 'length($5)>1 || length($6)>1 {print $2}' IBP_2.bim > redo_non_biallelic.txt
head redo_non_biallelic.txt
```

```bash
wc -l redo_non_biallelic.txt
```

```bash
plink --bfile IBP_2 --exclude redo_non_biallelic.txt --allow-no-sex --make-bed --out IBP_3
```
